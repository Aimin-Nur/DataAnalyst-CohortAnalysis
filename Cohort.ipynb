# Cohort Analysis

Sebuah project Data Analayst untuk mengetahui bagaimana pola pembelian pelanggan dalam periode waktu tertentu (Cohort Analysis).

Acquisition Cohort Analysis adalah jenis cohort analysis yang akan digunakan dalam project analisis kali ini, yang dimana saya akan mengambil insight Cohort Analysis berdasarkan tanggal atau waktu customer pertama kali melakukan transaksi dan kapan terakhir kali customer melakukan transaksi di platform e-commerce.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv', engine='python', encoding = 'ISO8859-1' )
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

### Reformat Timestamp

```python
import dateutil
from datetime import datetime as dt
from pytz import utc
```

```python
df['Datetime'] = df['InvoiceDate'].apply(lambda x : dateutil.parser.parse(x).timestamp())
df['Month'] = df['Datetime'].apply(lambda x : dt.fromtimestamp(x,utc).month)
df['Year'] = df['Datetime'].apply(lambda x : dt.fromtimestamp(x,utc).year)
```

```python
df.head()
```

### Cohort

```python
df['cohort'] = df.apply(lambda row : (row['Year']*100)+(row['Month']), axis=1)
cohorts = df.groupby('CustomerID')['cohort'].min().reset_index()
cohorts.columns = ['CustomerID', 'transaksi_pertama']
df = df.merge(cohorts, on='CustomerID', how='left')
```

```python
cohorts.head()
```

```python
df.head()
```

### Header setiap cohort

```python
headers = df['cohort'].value_counts().reset_index()
headers.columns = ['Cohorts', 'Count']
headers.head()
```

### Pivot Data

```python
headers = headers.sort_values(['Cohorts'])['Cohorts'].to_list()
headers
```

```python
df.head()
```

Penambahan Kolom "cohort_distance" untuk mengetahui berapa lama selisih customer melakukan transaksi pertama dengan terakhir kali melakukan transaksi.

```python
df['cohort_distance'] = df.apply(lambda row: (headers.index(row['cohort']) - headers.index(row['transaksi_pertama'])) if (row['transaksi_pertama'] != 0 and row['cohort'] != 0) else np.nan, axis=1)
```

```python
df.head()
```

```python
cohort_pivot = pd.pivot_table(df,
index='transaksi_pertama',
columns='cohort_distance',
values='CustomerID',
aggfunc=pd.Series.nunique)
```

```python
cohort_pivot
```

```python
cohort_pivot = cohort_pivot.div(cohort_pivot[0],axis=0)
```

```python
cohort_pivot
```

### Convert to Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig_dims = (12, 8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.heatmap(cohort_pivot, annot=True, fmt='.0%', mask=cohort_pivot.isnull(), ax=ax, square=True, linewidths=.5, cmap=sns.cubehelix_palette(rot=-.2))
plt.show()
```
